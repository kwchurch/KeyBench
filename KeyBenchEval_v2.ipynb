```python
def ExactMatchEval(gold, test):
    exactMatches = 0
    if len(test) == 0:
        return 0
    for kwd in test:
        if kwd in gold:
            exactMatches += 1
    return exactMatches
```

```python
def EditMatchEval(gold, test):
    editMatches = 0
    for kwd in gold:
        for kwd2 in test:
            dist_char = edit_distance(kwd.strip().lower(), kwd2)
            if dist_char >0 and dist_char <= 2:
                editMatches += 1
    return editMatches
```

```python
def ExactMatchEvalS(gold, test):
    porter = PorterStemmer()
    exactMatches = 0
    if len(test) == 0:
        return 0
    for kwd in test:
        for kwd2 in gold:
            if porter.stem(kwd) == porter.stem(kwd2):
                exactMatches += 1
    return exactMatches
```
